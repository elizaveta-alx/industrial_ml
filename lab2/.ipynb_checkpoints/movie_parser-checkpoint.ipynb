```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
from bs4 import BeautifulSoup
import pandas as pd
from tqdm import tqdm
from parsel import Selector
from imdb import Cinemagoer
import imdb
```

```python
#open url in a google chrome
def open_chrome_window(url):
    chromedrive_path = './chromedriver' 
    options = webdriver.ChromeOptions()
    options.add_argument('lang=en')
    driver = webdriver.Chrome(chromedrive_path, chrome_options=options)
    driver.get(url)
    driver.maximize_window()\
    return driver

#click a next page button
def click_next(driver):
    wait = WebDriverWait(driver, 10)
    next_button = wait.until(EC.element_to_be_clickable(
                           (By.XPATH, '//a[@class=\'lister-page-next next-page\']'))
                       )  
    next_button.click()
    
#get list of movies

#get movie short description and images using imdb package
def get_movie_descriprion_image(movie):    
    success_flag = False
    #while False:
    try:
        movie_obj = ia.search_movie(movie)[0]
        success_flag = True
    except IndexError:
        success_flag = False
        pass  
    id = movie_obj.getID()#movie_id
    movie_info = ia.get_movie(id)#movie info
    decription = movie_info['plot'][0]#short description
    image = movie_info['full-size cover url']#link to image
    return decription,image

def parse_movie_info(driver):
    movies = []
    i=0

    for i in tqdm(range(5)):
        response = BeautifulSoup(driver.page_source, 'html.parser') #get html page code
        title_list = response.find_all('h3', class_='lister-item-header') #create list with reviews data
        for r in tqdm(title_list):
            movie = []
            title = r.find('a').text
            movie.append(title)
            try:
                decription, image = get_movie_descriprion_image(title)
                movie.append(decription) 
                movie.append(image)
            except:
                pass
             
            #print(movie)
            movies.append(movie)   
        click_next(driver)
        i+=1
    return movies


def main():
    genres = pd.read_excel('./data/genres.xlsx')#list of all genres
    ia = imdb.IMDb()
    for ind in tqdm(genres.index):
        driver = open_chrome_window(genres['link'][ind])
        print(genres['title'][ind])
        movie_data = parse_movie_info(driver)
        driver.close()
        movie_data_df = pd.DataFrame(movie_data, columns =['title', 'description','image'])
        movie_data_df = movie_data_df[~movie_data_df['description'].isnull()].head(200)
        path = './movie_data/' + genres['title'][ind] + '.csv'
        movie_data_df.to_csv(path, encoding = 'utf-8')  #save df to csv
    return 0
```

```python
main()
```
