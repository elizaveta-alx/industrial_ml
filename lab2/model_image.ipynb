```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras.utils import to_categorical
from keras.preprocessing import image
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tqdm import tqdm
%matplotlib inline
```

```python
df_merged = pd.read_csv('movie_df_final.csv')
```

```python
df_merged.head()
```

```python
train_image = []

for i in tqdm(range(df_merged.shape[0])):
    img = keras.utils.load_img('./images/'+df_merged['image_name'][i],target_size=(400,400,3))
    img = keras.utils.img_to_array(img)
    img = img/255
    train_image.append(img)
    
```

```python
train_image
```

```python
X1 = np.array(train_image[0:1000])
X2 = np.array(train_image[1000:2000])
X3 = np.array(train_image[2000:3102])
X_test = np.array(train_image[3102:])
```

```python
X1[0]
```

```python
X_test.shape
```

```python
y1 = np.array(df_merged.drop(['title','description','image','genre','image_name'],axis=1).iloc[:1000,:])
y2 = np.array(df_merged.drop(['title','description','image','genre','image_name'],axis=1).iloc[1000:2000,:])
y3 = np.array(df_merged.drop(['title','description','image','genre','image_name'],axis=1).iloc[2000:3102,:])
y_test = np.array(df_merged.drop(['title','description','image','genre','image_name'],axis=1).iloc[3102:,:])
```

```python
y1.shape
```

```python
model = Sequential()
model.add(Conv2D(filters=16, kernel_size=(5, 5), activation="relu", input_shape=(400,400,3)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Conv2D(filters=32, kernel_size=(5, 5), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Conv2D(filters=64, kernel_size=(5, 5), activation="relu"))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Conv2D(filters=64, kernel_size=(5, 5), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(24, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
y_test.shape
```

```python
model.fit(X1, y1, epochs=10, validation_data=(X_test, y_test), batch_size=64)
```
