```python
import os
import glob
import pandas as pd
os.chdir(r'C:\Users\Elisaveta\Documents\ITMO\Industrial ML\movie_data')
```

```python
extension = 'csv'
#list of all files in directory
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
all_filenames
```

```python
#combine all files into one
combined_csv = pd.concat([pd.read_csv(f) for f in all_filenames ])
#export to csv
combined_csv.to_csv(r'C:\Users\Elisaveta\Documents\ITMO\Industrial ML\project\movie_data.csv', index=False, encoding='utf-8')
```

```python
combined_csv
```

```python
import string
#lower case and delete the punctuation
def text_preprocessing(description):
    description = description.lower()
    return description.translate(str.maketrans('', '', string.punctuation))

combined_csv['description'] = combined_csv['description'].apply(text_preprocessing)
combined_csv
```

```python
#group by title and and create column with all genres
movie_df = combined_csv.groupby('title')['genre'].apply(list).to_frame().reset_index()
#drop duplicates
df = combined_csv[['title','description','image']].drop_duplicates().merge(movie_df, how='right', on='title')
#save to csv
df.to_csv(r'C:\Users\Elisaveta\Documents\ITMO\Industrial ML\project\movie_data_merged_by_genre.csv', index=False, encoding='utf-8')
df
```

```python
import string

dir_path= r'C:\Users\Elisaveta\Documents\ITMO\Industrial ML\project'
new_dir_path = r'C:\Users\Elisaveta\Documents\ITMO\Industrial ML\project\images_new'

df = pd.read_csv(os.path.join(dir_path, 'movie_data_merged_by_genre.csv'))

def remove_punctuations(a_string):    
    return a_string.translate(str.maketrans('', '', string.punctuation))

df['index'] = df.index.astype(str)
df['genre_for_image_name'] = df['genre'].apply(remove_punctuations)
df['image_name'] = df['genre_for_image_name']+ '_' + df['index'] + '.jpg' 
df = df.drop(columns=['index','genre_for_image_name'])

df.to_csv(r'C:\Users\Elisaveta\Documents\ITMO\Industrial ML\project\movie_df_with_image_lables.csv')
```
